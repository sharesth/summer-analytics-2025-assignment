### 3. What is the most common origin for cars with horsepower > 100 and

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
import pandas as pd

# Load the CSV file into the DataFrame
cars_data = pd.read_csv("Cars.csv")

# Convert 'acceleration' to numeric, handling any non-numeric values
cars_data['acceleration'] = pd.to_numeric(cars_data['acceleration'], errors='coerce')

# Filter cars from Japan
japan_cars = cars_data[cars_data['origin'] == 'japan']

# Calculate the mean acceleration, rounded to 2 decimal places
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print(f"The mean acceleration of cars from Japan is {mean_acceleration} seconds.")
```

### 5. Which year had the highest average mpg?

```python
#your code here
import pandas as pd

# Load the CSV file into the DataFrame
cars_data = pd.read_csv("Cars.csv")

# Convert 'mpg' and 'model_year' to numeric, handling any non-numeric values
cars_data['mpg'] = pd.to_numeric(cars_data['mpg'], errors='coerce')
cars_data['model_year'] = pd.to_numeric(cars_data['model_year'], errors='coerce')

# Group by model_year and calculate the mean MPG for each year
yearly_avg_mpg = cars_data.groupby('model_year')['mpg'].mean()

# Find the year with the highest average MPG
highest_mpg_year = yearly_avg_mpg.idxmax()
highest_mpg_value = round(yearly_avg_mpg.max(), 2)

print(f"The year with the highest average MPG is {highest_mpg_year} with an average MPG of {highest_mpg_value}.")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
import pandas as pd

# Load the CSV file into the DataFrame
cars_data = pd.read_csv("Cars.csv")

# Convert relevant columns to numeric, handling any non-numeric values
cars_data['mpg'] = pd.to_numeric(cars_data['mpg'], errors='coerce')
cars_data['horsepower'] = pd.to_numeric(cars_data['horsepower'], errors='coerce')
cars_data['weight'] = pd.to_numeric(cars_data['weight'], errors='coerce')

# Calculate the median MPG
median_mpg = cars_data['mpg'].median()

# Filter cars with above-median MPG
above_median_mpg = cars_data[cars_data['mpg'] > median_mpg]

# Calculate horsepower-to-weight ratio
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Find the maximum hp_to_weight ratio
max_ratio = above_median_mpg['hp_to_weight'].max()

# Filter cars with the maximum hp_to_weight ratio
best_ratio_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]

# Print the car(s) with the best ratio
print("Car(s) with the best horsepower-to-weight ratio among cars with above-median MPG:")
for index, row in best_ratio_cars.iterrows():
    print(f"Car: {row['name']}, MPG: {row['mpg']}, Horsepower: {row['horsepower']}, Weight: {row['weight']}, HP-to-Weight Ratio: {round(max_ratio, 4)}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code hereimport pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into the DataFrame
cars_data = pd.read_csv("Cars.csv")

# Convert 'mpg' and 'model_year' to numeric, handling any non-numeric values
cars_data['mpg'] = pd.to_numeric(cars_data['mpg'], errors='coerce')
cars_data['model_year'] = pd.to_numeric(cars_data['model_year'], errors='coerce')

# Group by model_year and origin, then calculate the mean MPG
avg_mpg_by_year_origin = cars_data.groupby(['model_year', 'origin'])['mpg'].mean().unstack()



# Create the multi-line plot
plt.figure(figsize=(10, 6))
for origin in avg_mpg_by_year_origin.columns:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin], marker='o', label=origin.capitalize())

# Customize the plot
plt.title('Evolution of Average MPG Over Years by Origin', fontsize=14, pad=10)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.legend(title='Origin', fontsize=10)
plt.grid(True)
plt.xticks(avg_mpg_by_year_origin.index, rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Save the plot to a file
plt.savefig('mpg_evolution_by_origin.png')

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv('Cars.csv').dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])
for col in ['horsepower', 'weight', 'mpg']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df = df.dropna()

# Create visualization
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7,
    palette='viridis'
)

# Formatting
plt.title('Vehicle Performance Analysis: Horsepower vs Weight by Origin', pad=20)
plt.xlabel('Engine Power (Horsepower)', labelpad=15)
plt.ylabel('Vehicle Weight (lbs)', labelpad=15)
plt.grid(alpha=0.2)

# Legend handling
handles, labels = scatter.get_legend_handles_labels()
plt.legend(
    handles[0:6],
    ['Japan', 'Europe', 'USA', 'MPG (Size)'],
    title='Vehicle Origin & Efficiency',
    bbox_to_anchor=(1.25, 1),
    borderaxespad=0
)

plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Load and prepare data
df = pd.read_csv('Cars.csv').dropna(subset=['name', 'model_year', 'mpg'])
df['model_year'] = pd.to_numeric(df['model_year'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df = df.dropna(subset=['model_year', 'mpg'])

# Calculate consistency metrics
model_stats = df.groupby('name').agg(
    production_years=('model_year', 'nunique'),
    mpg_variation=('mpg', 'std')
).reset_index()

# Apply consistency criteria (multi-year production + low MPG variation)
consistent_cars = model_stats[
    (model_stats['production_years'] >= 2) &
    (model_stats['mpg_variation'] < 1.0)
].sort_values('mpg_variation')

consistent_cars
```
